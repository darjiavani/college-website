```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Input, Conv1D, GlobalMaxPooling1D, Dense

from sklearn.model_selection import train_test_split
from sklearn.metrics import (accuracy_score, confusion_matrix,
                             classification_report)
import matplotlib.pyplot as plt
```

```python
X=np.array([
[[3], [3], [3], [3], [3], [3], [3]],
 [[2], [2], [2], [2], [2], [2], [2]],
  [[7], [7], [7], [7], [7], [7], [7]],
   [[8], [8], [8], [8], [8], [8], [8]]
],dtype="float32")

y = np.array([1,1,0,0], dtype="float32")
```

```python
model = Sequential([
    Input(shape=(7,1)),
    Conv1D(3,4,activation="relu"),
    GlobalMaxPooling1D(),
    Dense(1,activation="sigmoid")
])

model.compile(
    loss="binary_crossentropy",
    optimizer="adam",
    metrics=["accuracy"]
)
```

```python
test1 = np.array([[[3], [3], [3], [3], [3], [3], [3]]], dtype="float32")
test2 = np.array([[[8], [8], [8], [8], [8],[8], [8]]],dtype="float32")
print("Efficient prob (low usage):",model.predict(test1, verbose=0))
print("Efficient prob (high usage):", model.predict(test2, verbose=0))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
print("train samples:", len(X_train), "test samples:", len(X_test))
```

```python
history = model.fit(
    X_train,
    y_train,
    epochs=100,
    verbose=1
    )
```

```python
model.summary()
```

```python
y_prob = model.predict(X_test, verbose=0).ravel()
print("predicted probabilities:", y_prob)

thresold = 0.5
y_pred = (y_prob > thresold).astype(int)
print("predicted classes:", y_pred)
print("true classes:", y_test)
```

```python
acc = accuracy_score(y_test, y_pred)
print("accuracy:", acc)

cm = confusion_matrix(y_test, y_pred)
print("confusion matrix:\n", cm)

cr = classification_report(y_test, y_pred, target_names=["not efficient", "efficient"])
print("classification report:\n", cr)
```
