```python
import cv2, numpy as np
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/CV.jpeg',0)

k = np.array([
    [1,4,6,4,1],
    [4,16,24,16,4],
    [6,24,36,24,6],
    [4,16,24,16,4],
    [1,4,6,4,1]
], np.float32)

k /= k.sum()
blur = cv2.filter2D(img,-1,k)
edges = cv2.Canny(blur,50,150)

print("Orignal Image")
cv2_imshow(img)
print("Blurred Image")
cv2_imshow(blur)
print("Canny edges")
cv2_imshow(edges)
```

```python
kernal = np.array([[0,-1,0],
                   [-1,5,-1]
                   [0,-1,0]])
sharp = cv2.filter2D(img,-1,kernal)
print("sharped")
cv2_imshow(sharp)
print("orignal")
cv2_imshow(img)

```
