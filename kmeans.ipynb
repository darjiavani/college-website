```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.DataFrame({
    "electricity_Kwh" : [120,130,125,300,320,310,70,75],
    "water_liters": [90,95,92,200,205,60,62,150]
})
```

```python
plt.scatter(df["electricity_Kwh"],df["water_liters"])
plt.xlabel("Electricity(Kwh)")
plt.ylabel("Water(liters)")
plt.title("Raw Data : Household Resource usages")
plt.show()
```

```python
KMeans_model = KMeans(n_clusters=3, random_state=42, n_init=10)
df["cluster"] = KMeans_model.fit_predict(df[["electricity_Kwh","water_liters"]])
```

```python
print("\nTable: Electricity, water, target(cluster)\n")
print(df[["electricity_Kwh","water_liters","cluster"]].to_string(index=False))
```

```python
new_Householdv= [[140,400]]
new_cluster = KMeans_model.predict(new_Householdv)[0]
print("\nNew household belongs to cluster:", new_cluster)
```
